```python
import numpy as np
import pandas as pd
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen=ImageDataGenerator(
    rescale=1/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
train_set=datagen.flow_from_directory(
    'dataset/training_set',
    batch_size=32,
    target_size=(64,64),
    class_mode='binary'
)
```

```python
datagen2=ImageDataGenerator(rescale=1/255)

test_set=datagen2.flow_from_directory(
    'dataset/test_set',
    batch_size=32,
    target_size=(64,64),
    class_mode='binary'
)
```

```python
cnn=tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(input_shape=[64,64,3],filters=32,kernel_size=3,activation='relu'))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=None))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=None))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(activation='relu',units=200))
```

```python
cnn.add(tf.keras.layers.Dense(activation='sigmoid',units=1))
```

```python
from tensorflow.keras.optimizers import SGD
cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(x=train_set, validation_data=test_set, epochs=30)
```
